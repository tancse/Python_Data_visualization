# PIE CHARTS

```python
#change the start angle
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 10, 55])
lab = ('Apples', 'Banana', 'Orange', 'Cherries')

plt.pie(y, labels= lab, startangle=90)
plt.show()
```

```python
#we can highlight some items in the pie chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 10, 55])
lab = ('Apples', 'Banana', 'Orange', 'Cherries')
ex = [0.2,0,0,0]

plt.pie(y, labels= lab, explode=ex, shadow=True)
plt.show()
```

```python
#colors
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 10, 55])
lab = ('Apples', 'Banana', 'Orange', 'Cherries')
ex = [0.2,0,0,0]
col = ['Blue','Cyan', 'Magenta','Green']

plt.pie(y, labels= lab, explode=ex, shadow=True, colors=col)
plt.show()
```

```python
#labelling
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 10, 55])
lab = ('Apples', 'Banana', 'Orange', 'Cherries')
ex = [0.2,0,0,0]

plt.pie(y, labels= lab, explode=ex, shadow=True)
plt.legend(title="Four Fruits:")
plt.show()
```

# STACK PLOT

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
studying = [7, 8, 6, 11, 7]
sleeping = [8, 10, 6, 10, 8]

plt.stackplot(days, studying, sleeping, colors= ['blue', 'orange'])
plt.xlabel('Days')
plt.ylabel('Hours')
plt.title('Stack Plot')
plt.legend(['Studying', 'Sleeping'])
plt.show()
```

# 3D PLOT

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 125]
# Creating the figure object
fig = plt.figure()
# keeping the projection = 3d
# creates the 3d plot
ax = plt.axes(projection = '3d')
ax.plot3D(z, y, x)
plt.show()
```

# Relational Plot

*   scatter plot

```python
import seaborn as sns
tips=sns.load_dataset("tips")
print(tips.head())
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style="sex", size="size")
```

```python
import seaborn as sns
tips=sns.load_dataset("tips")
print(tips.head())
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style="smoker", size="size", kind="line")
```

```python
#we can do scatter plot by including a linear regression model using lmplot()
import seaborn as sns
tips=sns.load_dataset("tips")
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

#DISTPLOT


*   histplot() - default one
*   kdeplot() - we can enable it using kind parameter also
*   ecdfplot() - same as kde
*   rugplot() - it is also a boolean parameter of distplot()



```python
#DISTPLOT
#histplot and kde
sns.displot(data=tips, x="total_bill", col="time")
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
# ecdf and rug plot
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker")
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker", rug=True)
```

```python
import plotly.graph_objects as go
import numpy as np

data = [[1,2,3,4,5],
	[3,4,5,6,7],
	[7,8,9,6,4],
	[3,7,2,4,2]]

fig = go.Figure(data =
	go.Contour(z = data))

fig.show()
```

```python
feature_x = np.arange(0, 50, 2)
feature_y = np.arange(0, 50, 3)

# Creating 2-D grid of features
[X, Y] = np.meshgrid(feature_x, feature_y)

Z = np.cos(X / 2) + np.sin(Y / 4)

fig = go.Figure(data =
	go.Contour(x = feature_x, y = feature_y, z = Z, colorscale="rainbow"))

fig.show()
```

```python
feature_x = np.arange(0, 50, 2)
feature_y = np.arange(0, 50, 3)

colorscale = colorscale = [[0, 'green'], [0.5, 'red'],
						[1.0, 'rgb(0, 0, 255)']]

[X, Y] = np.meshgrid(feature_x, feature_y)

Z = np.cos(X / 2) + np.sin(Y / 4)

fig = go.Figure(data =
	go.Contour(x = feature_x, y = feature_y, z = Z,
			colorscale = colorscale))

fig.show()
```
